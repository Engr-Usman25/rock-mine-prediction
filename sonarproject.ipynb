# Import Libaries 

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Preprocessing

```python
df= pd.read_csv('sonar data.csv')
```

# Separate the Label and Features

```python
X= df.drop(columns='R', axis=1)
Y= df['R']
```

```python
X   #all the data except the last column
```

```python
Y #the last column
```

# split into training and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Implement The Model

```python
md= LogisticRegression()
md.fit(X_train, Y_train)
```

# Model Evulation

```python
# Accuracy on training data
Y_train_pred= md.predict(X_train)
training_data_accuracy= accuracy_score(Y_train, Y_train_pred)
```

```python
training_data_accuracy
```

```python
# Accuracy on tEST data
Y_test_pred= md.predict(X_test)
test_data_accuracy= accuracy_score(Y_test, Y_test_pred)
```

```python
test_data_accuracy
```
